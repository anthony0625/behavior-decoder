```python
import numpy as np
import matplotlib.pyplot as plt
import math

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,8))
legs = 8

def conf(stdev, level=0.999, trials=100):
    if level==0.95:
        z = 1.960
    elif level==0.99:
        z=2.576
    elif level==0.999:
        z=3.291
    return z*stdev/math.sqrt(trials)

fig.suptitle("Correct/Incorrect 99.9% Confidence Interval")

ax[0].bar(["noise","s1"],[55.3333, 58.8889], yerr=[conf(23.5106),conf(11.8399)], capsize=legs)
ax[0].title.set_text("s1")

ax[1].bar(["noise","s2"],[55.3333, 80.0769], yerr=[conf(23.5106), conf(13.1020)], capsize=legs)
ax[1].title.set_text("s2")
```
